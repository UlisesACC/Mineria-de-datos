```python
# rango de las variables y valores que toman
data.describe().T
```

```python
for column in data.columns:
    if data[column].dtype == 'object':
        unique_values = data[column].unique()
        print(f"Columna: {column}, Valores únicos: {unique_values}")
    else:
        min_value = data[column].min()
        max_value = data[column].max()
        print(f"Columna: {column}, Rango: {min_value} - {max_value}")
```

```python
#columnas cuyo valor tienen numero 0 y cuantos ceros hay
for column in data.columns:
    zero_count = (data[column] == 0).sum()
    print(f"Columna: {column}, Cantidad de ceros: {zero_count}")
```
